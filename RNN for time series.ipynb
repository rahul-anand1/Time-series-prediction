```python
# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
def create_ts(start = '2001', n = 201, freq = 'M'):
    rng = pd.date_range(start=start, periods=n, freq=freq)
    ts = pd.Series(np.random.uniform(-18, 18, size=len(rng)), rng).cumsum()
    return ts
ts= create_ts(start = '2001', n = 192, freq = 'M')
ts.tail(5)
```

```python
ts = create_ts(start = '2001', n = 222)
```

```python
series = np.array(ts)
n_windows = 20
n_input = 1
n_output = 1
size_train = 201
```

```python
## Split data
train = series[:size_train]
test = series[size_train:]
```

```python
def create_batches(df, n_windows, n_input, output):
    ## Create x batches
    X_data = train[:size_train-1]
    X_batches = X_data.reshape(-1, n_windows, n_input)
    
    ## Create Y batches 
    Y_data = train[:size_train-1]
    Y_batches = Y_data.reshape(-1, n_windows, n_input)
    
    return X_batches, Y_batches
```

```python
X_batches, Y_batches = create_batches(df = train,
                                     n_windows = n_windows,
                                     n_input = n_input,
                                     output = n_output)
```

```python
X_test, Y_test = create_batches(df = test, n_windows = 20, n_input=1, output = 1)
```

```python
import tensorflow as tf
```

```python
X = tf.placeholder(tf.float32, [None, n_windows, n_input])
Y = tf.placeholder(tf.float32, [None, n_windows, n_input])
```

```python
r_neuron = 120   

## create model
basic_cell = tf.contrib.rnn.BasicRNNCell(num_units = r_neuron, activation = tf.nn.relu)
rnn_output, states = tf.nn.dynamic_rnn(basic_cell, X, dtype = tf.float32)
```

```python
stacked_rnn_output = tf.reshape(rnn_output, [-1, r_neuron])
stacked_output = tf.layers.dense(stacked_rnn_output, n_output)
outputs = tf.reshape(stacked_output, [-1, n_windows, n_output])
```

```python
learning_rate = 0.001

loss = tf.reduce_sum(tf.square(outputs - Y))
optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate)
training_op = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()
```

```python
iteration = 1500

with tf.Session() as sess:
    init.run()
    for iters in range(iteration):
        sess.run(training_op, feed_dict={X: X_batches, Y: Y_batches})
        if iters % 150 == 0:
            mse = loss.eval(feed_dict={X: X_batches, Y: Y_batches})
            print(iters, "\tMSE: ", mse)

    
    y_pred = sess.run(outputs, feed_dict={X: X_test})
```
